```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from PIL import Image

# Now you can use img_to_array
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
img = load_img('/content/drive/MyDrive/Surge(Microscopic Image Analysis)/Dataset/image2.png')
x = img_to_array(img)
x = x.reshape((1,)+x.shape)
i=0
for batch in datagen.flow(x,batch_size=1,
                          save_to_dir='/content/drive/MyDrive/Surge(Microscopic Image Analysis)/Dataset/additional', save_prefix='image', save_format='jpg'):
  i+=1;
  if(i>20):
    break
```
